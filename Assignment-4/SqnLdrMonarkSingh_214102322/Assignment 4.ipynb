```python
# Importing the dependencies
import numpy as np
from numpy import*
import matplotlib.pyplot as plt
from sympy import*
import sympy as sp
import cmath
```

### Question 1

```python
# Defining x[n]
N = 20           # Time Period
n = arange(-10,10,1)
x = np.zeros(N)

for i in range(N):
    if n[i] >= -2 and n[i] <= 2:
        x[i] = 1
        
# Plotting x[n]
x = np.tile(x,4)
n = np.arange(-3*10,5*10,1)

plt.figure(figsize=(14,6))
plt.plot(n,x,marker='o',color='maroon')
plt.xlabel('n',size=15)
plt.ylabel('x[n]',size=15)
plt.grid()
plt.show()
```

```python
# Computing the Discrete Time Fourier Series Coefficients:
C = np.zeros(N)

for n in range(N):
    for k in range(N):
        C[n] += x[k]*np.exp(-2j*np.pi*k*n/N)
C /= N     
        
mag = [0]*N
phi = [0]*N

# Computing the Magnitude and Phase
for i in range(N):
    mag[i],phi[i] = cmath.polar(C[i])
    
mag1 = np.zeros(N)
phi1 = np.zeros(N)

for i in range(N):
    mag1[i] = mag[i-int(N/2)]
    phi1[i] = phi[i-int(N/2)]
    
# Repeating the coefficients for three time periods
mag1 = np.tile(mag1,3)
phi1 = np.tile(phi1,3)
n = np.arange(-3*10,3*10,1)
```

```python
plt.figure(figsize=(16,8))

plt.subplot(211)
plt.plot(n,mag1,marker='o',color='teal')
plt.title('Fourier Series Magnitude Coefficients',size=15)
plt.xlabel('n',size=15)
plt.ylabel('Magnitude |Ck|',size=15)
plt.grid()

plt.subplot(212)
plt.plot(n,phi1,marker='o',color='olive')
plt.title('Fourier Series Phase Coefficients',size=15)
plt.xlabel('n',size=15)
plt.ylabel('Phase Ck',size=15)
plt.grid()

plt.tight_layout()
plt.show()
```

```python
# Defining the given function
n = np.arange(-5,6,1)
N = len(n)
x2 = np.zeros(N)

for i in range(N):
    if n[i]==-2 or n[i]==0 or n[i]==1:
        x2[i] = 1
    elif n[i]==-1 or n[i]==2:
        x2[i] = -1
    else:
        x2[i] = 0
        
# Plotting the given function
plt.figure(figsize=(10,5))
plt.stem(n,x2)
plt.xlabel('n',size=15)
plt.ylabel('x[n]',size=15)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Computing the Fourier Transform
N1 = 100
w = np.linspace(-np.pi,np.pi,N1) 
X_re = np.zeros(N1)
X_im = np.zeros(N1)

for i in range(N1):
    for n in range(int(-N/2),int(N/2)+1):
        X_re[i] += x2[n+int(N/2)]*cos(w[i]*n)    # Computing real part
        X_im[i] += -x2[n+int(N/2)]*sin(w[i]*n)   # Computing imaginary part

# w = np.linspace(0,2*np.pi,N1) 
# for i in range(N1):
#     for n in range(N):
#         X_re[i] += x2[n]*cos(w[i]*n)    # Computing real part
#         X_im[i] += -x2[n]*sin(w[i]*n)   # Computing imaginary part

# Computing Magnitude and Phase
mag = np.sqrt(X_re**2 + X_im**2)
ph = np.arctan(X_im/X_re)

mag1 = np.zeros(N1)
ph1 = np.zeros(N1)

for i in range(N1):
    mag1[i] = mag[i-int(N1/2)]
    ph1[i] = ph[i-int(N1/2)]
```

```python
# Plotting the Frequency Spectrum
plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
plt.plot(w,mag1, color='maroon')
plt.title('Fourier Transform: Magnitude Plot', size=15)
plt.xticks(np.arange(-np.pi, np.pi+np.pi/4, step=(np.pi/4)), ['-','-3/4','-/2','-/4','0','/4','/2','3/4',''])
plt.xlabel('', size=15)
plt.ylabel('|X()|', size=15)
plt.grid()

plt.subplot(1,2,2)
plt.plot(w,ph1, color='indigo')
plt.title('Fourier Transform: Phase plot', size=15)
plt.xticks(np.arange(-np.pi, np.pi+np.pi/4, step=(np.pi/4)), ['-','-3/4','-/2','-/4','0','/4','/2','3/4',''])
plt.xlabel('', size=15)
plt.ylabel('X()', size=15)
plt.grid()

plt.tight_layout()
plt.show()
```

```python
# Defining the given signal and impulse response
N = 100
n1 = np.linspace(0,10,N)
x3 = np.zeros(N)
h = np.ones(N)        # h[n] = u[n]

for i in range(N):
    x3[i] = 0.8**i
```

```python
'''
Defining the Convolution function
y[n] = x[n]*h[n]
N1 = length of x[n], N2 = length of h[n]
'''
def conv(x,h,N1,N2):
    y = [0] * (N1+N2-1)  # Declaring y as a zero array of length N1+N2-1
    for i in range(N1):
        for j in range(N2):
            y[i+j] += x[i]*h[j] 

    return y
```

```python
y = conv(x3,h,N,N)

n2 = np.linspace(0,20,2*N-1)

plt.figure(figsize=(10,5))
plt.plot(n2,y)
plt.title('LTI System Output: y[n] = x[n] * h[n]',size=15)
plt.xlabel('n',size=15)
plt.ylabel('y[n]',size=15)
plt.grid()
plt.show()
```

```python
# Computing the Fourier Transform
N2 = 100
w = np.linspace(0,2*np.pi,N2) 
Y = [0j]*N2
    
for i in range(N2):
    for n in range(N1):
        Y[i] += y[n]*np.exp(-1j*w[i]*n)  
        
r = [0]*N2
phi = [0]*N2

# Computing the Magnitude and Phase
for i in range(N2):
    r[i],phi[i] = cmath.polar(Y[i])
    
mag1 = np.zeros(N2)
ph1 = np.zeros(N2)

for i in range(N2):
    mag1[i] = r[i-int(N2/2)]
    ph1[i] = phi[i-int(N2/2)]
```

```python
# Plotting the Frequency Spectrum
plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
plt.plot(w,mag1, color='maroon')
plt.title('Fourier Transform: Magnitude Plot', size=15)
plt.xticks(np.arange(0, 2*np.pi+np.pi/4, step=(np.pi/4)), ['0','/4','/2','3/4','','5/4','6/4','7/4','2'])
plt.xlabel('', size=15)
plt.ylabel('|Y()|', size=15)
plt.grid()

plt.subplot(1,2,2)
plt.plot(w,ph1, color='indigo')
plt.title('Fourier Transform: Phase plot', size=15)
plt.xticks(np.arange(0, 2*np.pi+np.pi/4, step=(np.pi/4)), ['0','/4','/2','3/4','','5/4','6/4','7/4','2'])
plt.xlabel('', size=15)
plt.ylabel('Y()', size=15)
plt.grid()

plt.tight_layout()
plt.show()
```

### Discrete Fourier Transform

```python
# Computing and Plotting the Signals
t = np.linspace(0,10,64)
N = len(t)
f = np.zeros(N)

# Generating the signals 
for i in range(N):
    f[i] = np.exp(-t[i]**2/10)*(np.sin(2*t[i])+2*np.cos(4*t[i])+0.4*np.sin(t[i])*np.sin(10*t[i]))
    
# Plotting the signals
plt.figure(figsize=(10,5))

plt.plot(t,f, color='blue')
plt.xlabel('time (sec)',size=15)
plt.ylabel('f(t)',size=15)
plt.grid()
plt.show()
```

### Computing DFT

```python
# Defining a 2D array
N =64
rows = N 
cols = N

W = [[0 for i in range(rows)] for j in range(cols)]

# Computing the Fourier Coefficients using DFT
for i in range(rows):
    for k in range(cols):
        W[i][k] = np.exp(2j*np.pi*k/N)**i

Y = np.dot(W,f)

DFT_coeff = abs(Y)

plt.figure(figsize=(10,5))
plt.plot(DFT_coeff, color='teal')
plt.title('Fourier Coefficients', size=15)
plt.grid()
plt.show()
```

```python
# Defining the given Discrete Time sequence
n = np.arange(-20,21,1)
N = len(n)
x1 = np.zeros(N)
x2 = np.zeros(N)
x = np.zeros(N)

for i in range(N):
    if n[i]>=0:
        
        x1[i] = (-1/3)**n[i]
        
    else:
        x2[i] = (1/2)**n[i]
        
x = x1 - x2
```

### Z Transform of x[n]:
- X(z) = (z / (z + 1/3)) + (z / (z - 1/2))
- ROC: 1/3 < |z| < 1/2

```python
# Computing Z Transform:
N = 100
r = 0.4                          # |z| = 0.4
phi = np.linspace(0,2*np.pi,N)
z = [0j]*N
Xz = [0j]*N

for i in range(N):
    z[i] = r*np.exp(1j*phi[i])
    Xz[i] = (z[i]/(z[i]+1/3)) + (z[i]/(z[i]-1/2))

    
mag = [0]*N
phase = [0]*N

# Computing the Magnitude and Phase
for i in range(N):
    mag[i],phase[i] = cmath.polar(Xz[i])
    
mag1 = np.zeros(N)
phase1 = np.zeros(N)

for i in range(N):
    mag1[i] = mag[i-int(N/2)]
    phase1[i] = phase[i-int(N/2)]
```

```python
# Plotting x[n] and magnitude, phase spectrum of X(z)
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
plt.plot(n,x)
plt.title('Plot of x[n]', size=15)
plt.xlabel('n', size=15)
plt.ylabel('x[n]', size=15)
plt.grid()

plt.figure(figsize=(15,8))
plt.subplot(2,2,3)
plt.plot(phi,mag1, color='maroon')
plt.title('Z-Transform: Magnitude Plot', size=15)
plt.xticks(np.arange(0, 2*np.pi+np.pi/4, step=(np.pi/4)), ['0','/4','/2','3/4','','5/4','6/4','7/4','2'])
plt.xlabel('', size=15)
plt.ylabel('|X(z)|', size=15)
plt.grid()

plt.subplot(2,2,4)
plt.plot(phi,phase1, color='indigo')
plt.title('Z-Transform: Phase plot', size=15)
plt.xticks(np.arange(0, 2*np.pi+np.pi/4, step=(np.pi/4)), ['0','/4','/2','3/4','','5/4','6/4','7/4','2'])
plt.xlabel('', size=15)
plt.ylabel('X(z)', size=15)
plt.grid()

plt.tight_layout()
plt.show()
```
